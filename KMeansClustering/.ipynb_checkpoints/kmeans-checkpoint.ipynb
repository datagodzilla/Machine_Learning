## Assumptions
- Each cluster is generated by a multivariate normal distribution
- Total dataset can be regarded as a mixture of different distibutions
- Every cluster can be assigned 2 variables:
  - means
  - covariances
- If we know the correct cluster for every point, we can calculate the means and variances of the distribution for every cluster.
- The problems 


1. Initialize by assigning points randomly to clusters
2. Calculate means and covariances for every cluster
3. Calculate probabilities of cluster membership for every point and assign points to clusters with highest probability.
4. Goto 2
